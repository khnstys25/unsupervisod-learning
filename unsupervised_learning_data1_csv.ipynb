```python
import pandas as pd

# Membaca file CSV dari GitHub (versi raw) dan melewati baris error
Dataset = pd.read_csv(
    'https://raw.githubusercontent.com/jumadi-cloud/Fundamental-Python/main/Dataset/data1.csv',
    on_bad_lines='skip'
)

# Menampilkan nama-nama kolom
print(Dataset.keys())
```

```python
# Melakukan Coding disini
dataku = pd.DataFrame(Dataset)
dataku.head()
```

```python
import numpy as np

# Konversi ke data array
x = np.array(Dataset)
print(x)
```

```python
import matplotlib.pyplot as plt

# Menampilkan data dalam bentuk scatter plot
plt.scatter(x[:, 0], x[:, 1], label='True Position')

# Memberi label dan judul
plt.xlabel("X1")  # perbaikan: tanda kutip dibuka dan ditutup dengan benar
plt.ylabel("X2")
plt.title("Grafik Konsumen")
plt.legend()  # untuk menampilkan label
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Data dengan label dan angka
data = np.array([
    ['a', 11.432],
    ['b', 1.303],
    ['c', 99.906]
], dtype=object)

# Pisahkan fitur numerik (kolom ke-1)
X = data[:, 1].astype(float).reshape(-1, 1)  # reshape supaya jadi 2D array untuk KMeans

# Jalankan KMeans dengan k=2
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

# Output
print("Centroids:")
print(kmeans.cluster_centers_)

print("Labels:")
print(kmeans.labels_)

# Tampilkan hasil cluster beserta label asli
for label, cluster in zip(data[:, 0], kmeans.labels_):
    print(f"Data {label} masuk cluster {cluster}")
```
